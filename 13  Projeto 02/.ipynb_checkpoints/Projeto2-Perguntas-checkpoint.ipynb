## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
# solucao SQz
df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda']\
    .sum()\
    .nlargest(1)
```

```python
# solucao DSA
cidade_maior_venda = df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda']\
    .sum()\
    .idxmax()
print("Cidade com maior valor de venda para Office Suppliers:", cidade_maior_venda)
```

```python
# solucao DSA
df_dsa[df_dsa['Categoria'] == 'Office Supplies'].groupby('Cidade')['Valor_Venda']\
    .sum()\
    .sort_values(ascending=False)
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
# solucao SQz
vendas_segmentos = df_dsa.groupby('Segmento')['Valor_Venda']\
    .sum()
vendas_segmentos
```

```python
# solucao SQz
plt.pie(vendas_segmentos, labels='Segmento')
plt.show()
```

```python
# solucao DSA
vendas_segmentos = df_dsa.groupby('Segmento')['Valor_Venda']\
    .sum()\
    .reset_index()\
    .sort_values(by = 'Valor_Venda', ascending = False)
vendas_segmentos
```

```python
# solucao DSA
# funcao para converter os dados em valor absoluto
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# solucao DSA
plt.figure(figsize = (16,6))

# grafico de pizza
plt.pie(vendas_segmentos['Valor_Venda'],
       labels = vendas_segmentos['Segmento'],
       autopct = autopct_format(vendas_segmentos['Valor_Venda']),
       startangle = 90)

# limpa o circulo central
centre_circle = plt.Circle((0,0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# labels e anotacoes
plt.annotate(text = 'Total de vendas: ' + '$ ' + str(int(sum(vendas_segmentos['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de vendas por segmento')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
# Solucao DSA

# cria uma nova coluna de acordo com a regra definida
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
df_dsa.head()
```

```python
# Solucao DSA

# total por cada valor da variavel
df_dsa['Desconto']\
    .value_counts()\
    .reset_index()
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Solucao DSA

# Valor simulado da venda, menos o desconto
df_dsa['Valor_Venda_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
df_dsa.head()
```

```python
# Solucao DSA

# filtrando vendas antes do desconto de 15%
vendas_antes_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
vendas_antes_desconto
```

```python
# Solucao DSA

# filtrando vendas depois do desconto de 15%
vendas_depois_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Desconto']
vendas_depois_desconto
```

```python
# Solucao DSA

# media das vendas antes do desconto de 15%
media_vendas_antes_desconto = vendas_antes_desconto.mean()
media_vendas_antes_desconto
```

```python
# Solucao DSA

# media das vendas depois do desconto de 15%
media_vendas_depois_desconto = vendas_depois_desconto.mean()
media_vendas_depois_desconto
```

```python
# Solucao DSA

print('Média das vendas antes do desconto de 15%: ', round(media_vendas_antes_desconto, 2))
print('Média das vendas depois do desconto de 15%: ', round(media_vendas_depois_desconto, 2))
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
# Solucao DSA

# extracao do mes
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
df_dsa
```

```python
# Solucao DSA

# estatistica de agregacao
media_anomes = df_dsa.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda']\
    .agg([np.sum, np.mean, np.median])
media_anomes
```

```python
# Solucao DSA

# extracao niveis
anos = media_anomes.index.get_level_values(0)
meses = media_anomes.index.get_level_values(1)
segmentos = media_anomes.index.get_level_values(2)
anos
```

```python
# Solucao DSA
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = media_anomes,
                  y = 'mean',
                  x = meses,
                  hue = segmentos,
                  col = anos,
                  col_wrap = 4)
plt.show()
```

```python
# Solucao DSA
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = media_anomes,
                  y = 'sum',
                  x = meses,
                  hue = segmentos,
                  col = anos,
                  col_wrap = 4)
plt.show()
```

```python
# Solucao DSA
plt.figure(figsize = (12,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = media_anomes,
                  y = 'median',
                  x = meses,
                  hue = segmentos,
                  col = anos,
                  col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
# Solucao DSA

# agrupamos por categoria e subcategoria e calculamos a soma somente para variaveis numericas
vendas_top12 = df_dsa.groupby(['Categoria', 'SubCategoria'])\
    .sum(numeric_only = True)\
    .sort_values('Valor_Venda', ascending = False)\
    .head(12)
vendas_top12
```

```python
# Solucao DSA

# convertemos valor_venda em inteiro e classificamos por categoria
vendas_top12 = vendas_top12[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
vendas_top12
```

```python
# Solucao DSA

# criamos outro dataframe com totais por categoria
vendas_top12_cat = vendas_top12.groupby('Categoria')\
    .sum(numeric_only = True)\
    .reset_index()
vendas_top12_cat
```

```python
# Solucao DSA
# cores
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Solucao DSA

fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(vendas_top12_cat['Valor_Venda'], 
            radius = 1,
            labels = vendas_top12_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(vendas_top12['Valor_Venda'],
            radius = 0.9,
            labels = vendas_top12['SubCategoria'],
            autopct = autopct_format(vendas_top12['Valor_Venda']),
            colors = cores_subcategorias, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(vendas_top12['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
